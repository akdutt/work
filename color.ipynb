```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:/Users/MAX/Desktop/office/colours.jpg")
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(RGB)
```

```python
HSV=cv2.cvtColor(RGB,cv2.COLOR_RGB2HSV)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(HSV)
```

```python
lower= np.array([25,150,50])
upper =np.array([35,255,255])
```

```python
detect = cv2.inRange(HSV,lower,upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(detect)
```
